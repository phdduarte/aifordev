```python
import pandas as pd
import requests
```

```python
# Carregar o arquivo CSV
df = pd.read_csv('./all.csv')
```

```python
comments = df.iloc[:, 1].tolist()
```

```python
print(comments[:5])
```

```python
# Inicializar listas para armazenar as respostas da API
success_list = []
flagged_list = []
method_list = []
```

```python
# Fazer a requisição POST para a API para cada comentário
url = "http://localhost:8787/api/concursos/moderate/comments"
for index, comment in enumerate(comments):
    if index >= 81531:
        break
    try:
        response = requests.post(url, json={"comment": comment})
        response.raise_for_status()  # Levanta um erro para códigos de status HTTP que indicam erro
        data = response.json()

        # Armazenar as respostas da API nas listas
        success_list.append(data.get('success', None))
        flagged_list.append(data.get('flagged', None))
        method_list.append(data.get('method', None))
        print(f"{index}: {data.get('success', None)}\n")
    except requests.exceptions.RequestException as e:
        # Logar o erro e continuar o loop
        print(f"Erro na requisição para o comentário: {comment}\nErro: {e}")
        success_list.append(None)
        flagged_list.append(None)
        method_list.append(None)
```

```python
# Adicionar as novas colunas ao DataFrame
df['success'] = success_list
df['flagged'] = flagged_list
df['method'] = method_list
```

```python
# Salvar o DataFrame atualizado em um novo arquivo CSV
df.to_csv('./test-comentarios-updated.csv', index=False)
```
